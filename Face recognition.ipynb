# Flattening

```python
flatten_size_gal= image_size_gal[0]*image_size_gal[1]
flatten_size_gal
```

```python
for i in range(len_of_images_gal):
    gal_grey[i]=np.ndarray.flatten(gal_grey[i]).reshape(flatten_size_gal,1)
```

```python
gal_grey=np.dstack(gal_grey)
```

```python
gal_grey=np.rollaxis(gal_grey,axis=2,start=0)
gal_grey.shape
```

```python
gal_grey=gal_grey.reshape(len_of_images_gal,flatten_size_gal)
```

```python
gal_grey.shape
```

```python
flatten_size_emma = image_size_emma[0]*image_size_emma[1]
flatten_size_emma
```

```python
for i in range(len_of_images_emma):
    emma_grey[i]=np.ndarray.flatten(emma_grey[i]).reshape(flatten_size_emma,1)
```

```python
emma_grey=np.dstack(emma_grey)
```

```python
emma_grey=np.rollaxis(emma_grey,axis=2,start=0)
emma_grey.shape
```

```python
emma_grey=emma_grey.reshape(len_of_images_emma,flatten_size_emma)
```

```python
emma_grey.shape
```

```python
flatten_size_natalia = image_size_natalia[0]*image_size_natalia[1]
flatten_size_natalia 
```

```python
for i in range(len_of_images_natalia):
    natalia_grey[i]=np.ndarray.flatten(natalia_grey[i]).reshape(flatten_size_natalia,1)
```

```python
natalia_grey=np.dstack(natalia_grey)
```

```python
natalia_grey=np.rollaxis(natalia_grey,axis=2,start=0)
natalia_grey.shape
```

```python
natalia_grey=natalia_grey.reshape(len_of_images_natalia,flatten_size_natalia)
```

# Module_selection

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# SUPPORT VECTOR MACHINE

```python
from sklearn import svm
```

```python
clf=svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
y_pred
```

# prediction accuracy

```python
from sklearn import metrics
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```
